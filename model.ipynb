```python
from sklearn.model_selection import train_test_split
import pandas as pd
df=pd.read_csv("dataset.csv")

train, test = train_test_split(df, test_size=0.2)
print(len(train))
print(len(test))
```

```python
import numpy as np
from simpletransformers.classification import ClassificationModel, ClassificationArgs


train_df=pd.DataFrame(train)
train_df.columns=['Review_text','sentiment']

eval_df=pd.DataFrame(test)
eval_df.columns=['Review_text','sentiment']

print(train_df)
print(eval_df)





```

```python
model_args=ClassificationArgs(num_train_epochs=3,learning_rate=4e-5,save_eval_checkpoints=True,
                              overwrite_output_dir=True,
                              save_model_every_epoch=True,
                              best_model_dir=o'utputs/best_model')

model=ClassificationModel(
    'bert',
    'bert-base-uncased',
    num_labels=13,
    args=model_args,
    use_cuda=False
  
)
```

```python
model.train_model(train_df)
```

```python
model.train_model(eval_df)
```

```python
result,model_outputs,wrong_predictions=model.eval_model(eval_df)
```
